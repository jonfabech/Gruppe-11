```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re

url = 'https://www.airbnb.dk/s/Copenhagen--Denmark/homes?allow_override%5B%5D= '

path2gecko ='/Users/Majsa/Downloads/geckodriver'# define path to y our geckodriver
browser = webdriver.Firefox(executable_path=path2gecko)


browser.get(url)
text = browser.page_source

soup = BeautifulSoup(text, "lxml")
```

- Find enkelte lejligheder og uddrag 1) link og 2) titel

```python
lejligheder = soup.findAll("div",{'class':"container_j4k9t6-o_O-column-sm-1_zdxht7-o_O-column-md-2_3504zm-o_O-column-lg-2_14svpwf-o_O-column-xl-3_19vi8ax"}) # locate articles using the findall function. 
lejligheder_navn = soup.findAll("span",{'class':'accessible_1m8bb6v'}) # locate articles using the findall function. 

baseurl = 'https://www.airbnb.dk'
links = []
navne = []

for lej in lejligheder:
    link = baseurl+lej.a['href']
    links.append(link)

for lej in lejligheder_navn:
    lej_t=lej.text
    navne.append(lej_t)
    
navne = navne[0::2]
```

```python
import os,time,codecs

pris = []
kapacitet = []
badev = []
sovev = []
senge = []
vtype = []

for link in links:
    time.sleep(3) # wait a little
    browser.get(link) #jeg henter de respektive links i variablen "links"
    text_i = browser.page_source #jeg aktiverer geckodriveren(?)
    soup = BeautifulSoup(text_i, "lxml") #jeg laver html´en om til en smuk suppe
    
    #jeg specificerer den del hvor prisen ligger
    enkelt_pris = soup.findAll("span",{'class':'text_5mbkop-o_O-size_large_16mhv7y-o_O-weight_bold_153t78d-o_O-color_inverse_1lslapz-o_O-inline_g86r3e'}) # locate articles using the findall function. 
    #jeg specificerer den del hvor kapaciteten ligger
    enkelt_plads = soup.findAll("div",{'class':'bottom-spacing-2'})
    
    #jeg henter fra-prisen på hver enkelt lejlighed
    for enk in enkelt_pris:
        pris_t=enk.text
        pris.append(pris_t)
    
    #jeg henter kapaciteten. For ikke at få "kapacitet" med, bruger jeg regex til kun at udtrække tal 
    #til sidst appender jeg tallet til variablen kapacitet
    #Det viser sig, at man muligvis ikke kan gøre det med index. 
    #badeværelse er nemlig ikke altid [1]
    kapacitet_ = enkelt_plads[0].text
    kapacitet_ = re.findall('\d+', kapacitet_)
    kapacitet.append(kapacitet_)
    
    #samme som for kapacitet
    badev_ = enkelt_plads[1].text
    badev_ = re.findall('\d+', badev_)
    badev.append(badev_)
    
    #samme som for kapacitet
    sovev_ = enkelt_plads[2].text
    sovev_ = re.findall('\d+', sovev_)
    sovev.append(sovev_)
    
    #samme som for kapacitet
    senge_ = enkelt_plads[3].text
    senge_ = re.findall('\d+', senge_)
    senge.append(senge_)
    
    #Jeg henter værelsestypen. Jeg splitter variablen ved ":" for ikke at få "Værelsestype" med.
    #Jeg stripper " " for at fjerne mellemrummet foran værelsestypen.
    #Indekseringen virker ikke i for loopet
    #vtype_ = enkelt_plads[7].text
    #vtype_ = ((vtype_.split(":"))[1]).strip(" ")
    #vtype.append(vtype_)    
```
